```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 12 ## 10-armed bandits
n_p = 2000 ## No of simulations

q_star = np.random.normal(0, 1, (n_p,n))
arms = [0] * n ## set of actions(arms to be pulled)

for i in range(12):
    arms[i] = np.random.normal(q_star[0, i], 1, 2000)
```

```python
plt.figure(figsize=(12,8))
plt.ylabel('Rewards(Normal) distribution')
plt.xlabel('Set of Actions')
plt.xticks(range(1,13))
plt.yticks(np.arange(-5,5,0.5))

plt.violinplot(arms, positions=range(1,13), showmedians=True)
plt.show()
```
